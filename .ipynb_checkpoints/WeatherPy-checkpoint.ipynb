Summary:
1. The temperature is related to the latitude. The closer the latitude is to zero , the higher the temperature.
2. The cloudiness, humidity and wind speed is not that related to the latitude.
3. Most city's wind speed is below 20 mph.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import datetime
from scipy.stats import linregress


# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Create basic url
url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={weather_api_key}"
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# Create the lists to store data
city_list = []
country = []
date = []
cloud = []
hum = []
lat = []
lng = []
temp = []
wind = []

print("Beginning Data Retrieval")
print("------------------------")

# Start data retrieval with loop

nrecord = 0
nset = 1

for city in cities:
    q_url = url + "&q=" + city
    response = requests.get(q_url).json()
    if nrecord < 50:
        nrecord += 1
    else:
        nset += 1
        nrecord = 1
        
    print(f"Processing Record {nrecord} of {nset} | {city}")
    
    try:
        city_list.append(response["name"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        cloud.append(response["clouds"]["all"])
        hum.append(response["main"]["humidity"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        temp.append(response["main"]["temp"])
        wind.append(response["wind"]["speed"])
        
    except:
        print("City not found! Skipping...")
        pass
    
print("-----------------------")
print("Data Retrieval Complete")
print("-----------------------")    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Display the dataframe
weather = {"city" : city_list,
           "country" : country,
           "date" : date,
           "cloudiness" : cloud,
           "humidity" : hum,
           "latitude" : lat,
           "longitude" : lng,
           "temperature" : temp,
           "wind speed": wind}
weather_df = pd.DataFrame(weather)
weather_df.head()
```

```python
# Export to .csv
weather_df.to_csv(output_data_file, index=False)
```

```python
# Get current date
d = datetime.datetime.today()
today_date = d.strftime('%m/%d/%Y')
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
n_df = weather_df.loc[weather_df["latitude"] >= 0]
s_df = weather_df.loc[weather_df["latitude"] < 0]
```

####  Northern Hemisphere - Temp vs. Latitude Linear Regression

```python
x_values = n_df["latitude"]
y_values = n_df["temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5,15),fontsize=15,color="red")
plt.title("Northern Hemisphere - Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.grid()
print(f"The r-squared is {rvalue**2}")
plt.savefig("output_data/NHemisphere_Latitude_Temperature_Plot.png")
```

As r-squared value is big enough, the model fits the data very well. The higher the latitude of northern hemisphere is, that city is getting colder.

####  Southern Hemisphere - Temp vs. Latitude Linear Regression

```python
x_values = s_df["latitude"]
y_values = s_df["temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(-30,50),fontsize=15,color="red")
plt.title("Southern Hemisphere - Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.grid()
print(f"The r-squared is {rvalue**2}")
plt.savefig("output_data/SHemisphere_Latitude_Temperature_Plot.png")
```

The r-squared value is 0.52, which is not very high, so the model fits the data ok but not great. Based on this model, the higher the latitude of southern hemisphere is, that city is getting hotter.
